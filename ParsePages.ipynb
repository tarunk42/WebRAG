```python
import json

# Function to parse a JSON file and extract text
def parse_json_file(file_path):
    try:
        # Load the JSON file
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        
        # Extract text
        # Assuming the JSON contains a key "text" that holds the text content
        extracted_text = []
        if isinstance(data, dict):  # JSON is an object
            extracted_text.append(data.get("text", ""))
        elif isinstance(data, list):  # JSON is an array
            for item in data:
                if isinstance(item, dict):
                    extracted_text.append(item.get("text", ""))
        
        # Return the extracted text
        return extracted_text
    
    except Exception as e:
        print(f"Error parsing the JSON file: {e}")
        return []

# Example Usage
file_path = './page_test13.json'  # Replace with your JSON file path
text_data = parse_json_file(file_path)

# Print extracted text
print("Extracted Text:")
for text in text_data:
    print(text)
```
